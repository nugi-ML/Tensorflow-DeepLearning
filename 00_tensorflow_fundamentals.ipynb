```python
# Can do the same with transpose
X, tf.transpose(X), tf.reshape(X, shape=(2, 3))
```

```python
# try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

## Aggregating tensors

Aggregating tensors = condesing them from multiple values down to a smaller amount of values

```python
# Get the absolute values
D = tf.constant([-7, -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let's go through the following forms of aggregation:

* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor woth values between 0 and of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E. shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Finde the sum
tf.reduce_sum(E)
```

```python
# Find the variance of our tensor, we need access to tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

```python
tf.math.reduce_std(E)
```

```python
# Find the standard deviation
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

## Find the positional maximum and minimun

```python
# Create a new tensor for finding positional minimun and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# Indec on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum using the positional minimum index
F[tf.argmin(F)]
```
